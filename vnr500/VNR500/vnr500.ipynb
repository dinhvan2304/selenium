```python
import pandas as pd
import numpy as np

names = ['Company', 'CEO', 'MST', 'NGANH_NGHE']
data1 = pd.read_csv('vnr_profit_500_1.csv',names= names,dtype = {'MST':np.str})
# data1['Type'] = 1
data2 = pd.read_csv('vnr_profit_500.csv',names=names ,dtype = {'MST':np.str})
# data2['Type'] = '2'
data2 = data2.drop_duplicates(subset = ['MST'])
data2 = data2.reset_index(drop = True)
data = pd.concat([data1,data2],ignore_index = True)
data = data.drop_duplicates(subset=['MST'])
data = data.reset_index(drop = True)
data
```

```python
import pandas as pd
import numpy as np
data3 = pd.read_csv('Result_total_vietstock.csv',dtype = {'mst': np.str})
data3 = data3.dropna(subset = ['mst'])
data3['mst'] = data3['mst'].astype('string')
data3['mst'] = data3['mst'].str.strip()
data3['mst'] = data3['mst'].str.replace(' ','')
data3['mst'] = data3['mst'].str[:10]
data3['mst'] = data3['mst'].str.replace('.','')
# data3['mst'] = data3['mst'].str.replace('','')
data3['mst'] = data3['mst'].apply(lambda x: x if len(x) == 10 else '0'+ x)
data3.loc[data3['doanh_thu_2020'].isna() == True].count()
data3
```

```python
d2020 = dict(zip(data3.mst,data3.doanh_thu_2020))
d2021 = dict(zip(data3.mst,data3.doanh_thu_2021))
data['doanh_thu_2020'] = data['MST'].map(d2020) 
data['doanh_thu_2021'] = data['MST'].map(d2021) 
# data.loc[data['doanh_thu_2020'].isna() == True]
# data.to_csv('VNR500_doanh_thu.csv')
data
```
