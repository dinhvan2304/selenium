```python
mstcn_name = pd.read_csv('data/mstcn_name.txt')

def remove_bracket(x):
    names = x.strip()
    if "(" in names:
        index_bracket = names.index("(")
        name = names[:index_bracket]
    elif "-" in names:
        index_dash = names.index("-")
        name = names[:index_dash]        
    else:
        name = names

    return name


mstcn_name['Name'] = mstcn_name['Name'].apply(lambda x: remove_bracket(x))
mstcn_name = mstcn_name[~mstcn_name['Name'].str.contains("~")]
mstcn_name.to_csv('data/mstcn_name.txt', index=False)
```

```python
mstcn_name = pd.read_csv('data/mstcn_name.txt')

def remove_multi_space(x):
    if "        " in unidecode.unidecode(x):
        x = x.replace("        ", " ")
    
    return x

mstcn_name['Name'] = mstcn_name['Name'].apply(lambda x: remove_multi_space(x))

mstcn_name.to_csv('data/mstcn_name.txt', index=False)
```
